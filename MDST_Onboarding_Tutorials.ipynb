**Feature engineering justification:**

Make a new column called SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT, which normalizes the difference between populated states and unpopulated states by taking the number of students into consideration when looking at SUPPORT_SERVICES_EXPENDITURE.

Make several new columns, which are the log-transformation of TOTAL_REVENUE, FEDERAL_REVENUE, STATE_REVENUE, LOCAL_REVENUE. For example, when looking at the TOTAL_REVENUE column, California seems like an outlier on the raw scale. We take log-transformation to minimize this phenomenon (Further explanation below).

```python
# Total revenue are right-skewed.
# Therefore, extremely large values appear as outliers on the raw scale
# We can minimize this phenomenon by taking the log-transformation
df.TOTAL_REVENUE.plot.hist(title="Total Revenue", edgecolor="black")
```

```python
# Making a new column called LOG_TOTAL_REVENUE
df['LOG_TOTAL_REVENUE'] = np.log(df['TOTAL_REVENUE'])

# Plot the distribution of LOG_TOTAL_REVENUE
df.LOG_TOTAL_REVENUE.plot.hist(title="Log(Total Revenue)", edgecolor="black")

# We can see that by taking the log-transformation
# total revenue now has a distribution more similar to a normal Gausian distribution
```

```python
# Same thing, taking the log-transformation of FEDERAL_REVENUE, STATE_REVENUE, LOCAL_REVENUE
df['LOG_FEDERAL_REVENUE'] = np.log(df['FEDERAL_REVENUE'])
df['LOG_STATE_REVENUE'] = np.log(df['STATE_REVENUE'])
df['LOG_LOCAL_REVENUE'] = np.log(df['LOCAL_REVENUE'])
```

```python
# In this case, we know our data is on the state level and also longitudinal (over time).

# This format introduces some complications.
# For example, the state of California will obviously spend more than New Jersey
# becuase they have more people... how can we account for this?

# let's create a new column which represents expenditure per student
df['SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT'] = df['SUPPORT_SERVICES_EXPENDITURE'] / df['ENROLL_ALL']
```

## Data Creation

```python
from sklearn.model_selection import train_test_split
# X is commonly used to denote the input data
# y is used for the response / output data
X = df[['SUPPORT_SERVICES_EXPENDITURE_PER_STUDENT','AVG_READING_4_SCORE']].dropna()
y = df.loc[X.index]['AVG_MATH_4_SCORE']
```

```python
# We also need to make sure there is no NaN in y
# This time, we will fill the NaN with the median of y
# We prefer median to mean because EDA reveals that the response variable is left-skewed. Therefore, the mean may not represent the data very well
y.fillna(y.median(), inplace=True)
```

```python
# the test_size parameter defines what % of data is set aside for testing, 70 / 30 and 80 / 20 split are both typical
# we don't have a huge data set but we still want to have a decently sized testing set
# so we are using a 70 / 30 train / test split.
# setting random_state explicitly ensures that I get the same results each time I run the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3, random_state=0)
```

## Prediction

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
# fit is essentially the word sklearn uses for training
model.fit(X_train, y_train)
```

```python
# You can see the intercepts and coefficients the model generates
print(model.intercept_)
print(model.coef_)
```

## Evaluation

```python
# R^2 value describes how well a linear model fits the data
# It ranges between 0 and 1
# There are many caveats to R^2 but it is a good starting point
model.score(X_test, y_test)
```

```python
# mean error
np.mean(model.predict(X_test)-y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
```

```python
loan_data.head()
```

```python
# Dependent variables are everything except for loan_id and loan_status
X = loan_data.drop(['loan_id', 'loan_status'], axis=1)
Y = loan_data['loan_status']

# Initiate training and testing datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state = 42)

# Initiate an instance of sklearn.DecisionTree
DecisionTree = DecisionTreeClassifier(random_state = 42)
```

## Model Training

```python
# Train model on training data
DecisionTree_fit = DecisionTree.fit(X_train, Y_train)
```

##Tree Pruning

**Decision tree** are notorious for being **overfit** to the **Training Dataset**, and there are a lot of parameters, like <mark>*max_depth*</mark> and <mark>*min_samples*</mark>, that are designed to reduce overfitting. However, pruning a tree with a **cost complexity pruning** can simplify the whole process of finding a smaller tree that improves the accuracy with the **Testing Dataset**

Pruning the tree is all about fiding the right value for the pruning parameter, <mark>*alpha*</mark>, which controls how little or how much pruing happens. One way to find the optimal value for *alpha* is to plot the accuracy of the tree as a function of different values. We'll do this for both **Training Dataset** and **Testing Dataset**

First, let's extract <mark>*alpha*</mark> that are available for this tree and build a pruned tree for each value for alpha. **Note:** We omit the maximum value for <mark>*alpha*</mark> with <mark>*ccp_alphas = ccp_alphas[:-1]*</mark> because it would prune all leaves, leaving us with only a root instead of a tree.

```python
# Cost complexity pruning - Because decision trees are notorious for overfitting
# Once we found alpha, we need to exclude the largest value, because if not, every leave will be pruned

path = DecisionTree_fit.cost_complexity_pruning_path(X_train, Y_train) # Find alpha
ccp_alphas = path.ccp_alphas # extract alphas, ccp = cost complexity pruning
ccp_alphas = ccp_alphas[:-1] # remove the largest alpha

DecitionTrees = [] # Create array to store decision trees into

# Create one decision tree per value for alpha and store it in the array
for alpha in ccp_alphas:
    DecisionTree_fit = DecisionTreeClassifier(random_state = 0, ccp_alpha = alpha)
    DecisionTree_fit.fit(X_train, Y_train)
    DecitionTrees.append(DecisionTree_fit)
```

```python
# Compute scores for each alpha
train_scores = [DT.score(X_train, Y_train) for DT in DecitionTrees]
test_scores = [DT.score(X_test, Y_test) for DT in DecitionTrees]

fig, ax = plt.subplots()
ax.set_xlabel("alpha")
ax.set_ylabel("accuracy")
ax.set_title("Accuracy vs Alpha for training and testing sets")
ax.plot(ccp_alphas, train_scores, marker = "o", label = "train", drawstyle="steps-post")
ax.plot(ccp_alphas, test_scores, marker = "o", label = "test", drawstyle="steps-post")
ax.legend()
plt.show()
```

## Cross Validation

```python
ccp_alphas
```

The graph we drew suggested one value of <mark>*alpha*</mark>, **0.00030736**, which is the alpha value corresponding to the point prior to a significant drop in accuracy of both training and testing.

First, let's demonstrate that different training and testing datasets result in trees with different accuracies when we set <mark>*ccp_alpha=0.00030736*</mark>. we will do this using the <mark>*cross_val_score()*</mark> function to generate different training and testing datasets and then train and test the tree with those datasets.

```python
DT = DecisionTreeClassifier(random_state = 42, ccp_alpha = 0.445)

## Now use 5-fold corss validation, create 5 different training and testing datasets that
## are then used to train and test the tree
## NOTEL: We use 5-fold because we don't have tons of data...

scores = cross_val_score(DT, X_train, Y_train, cv = 5)
df = pd.DataFrame(data={'tree': range(5), 'accuracy': scores})

df.plot(x="tree", y="accuracy", marker="o", linestyle="--")
```

```python
## create an array to store the results of each fold during cross validation
alpha_loop_values = []

## for each candidate value for alpha, we will run 5-fold cross validation
## then we will score the mean and standard deviation of the score (the accuracy) for each call
## to cross_val_score in alpha_loop_values...

for ccp_alpha in ccp_alphas:
    DT = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    scores = cross_val_score(DT, X_train, Y_train, cv=5)
    alpha_loop_values.append([ccp_alpha, np.mean(scores), np.std(scores)])
```

```python
## now we can draw a graph of the means and standard deviations of the scores
## for each candidate value alpha
alpha_results = pd.DataFrame(alpha_loop_values, columns = ['alpha', 'mean_accuracy', 'std'])

alpha_results.plot(x='alpha',
                  y='mean_accuracy',
                  yerr="std",
                  marker="o",
                  linestyle="--")
```
